```python
import numpy as np
```

```python
def FF(item_l,L):
    l = list(item_l.values())
    n = len(l)
    m = np.ceil(sum(l)/L).astype(int)
    N = list(item_l.keys()) #itens
    M = list(range(m)) #bins
    sobra = {j:L for j in M} #Espaço que resta em cada bin
    usados = {} #Bins usados
    minimo = min(l)
    for item in N:
        i = 0
        for caixa in M:
            if item_l[item]<= sobra[caixa]:
                try:
                    usados[caixa].append(item)
                    sobra[caixa] -= item_l[item]
                    if sobra[caixa]<minimo: #Se na caixa não couber mais o menor item, então a deletamos
                        del M[M.index(caixa)]
                    break
                except:
                    usados[caixa] = [item]
                    sobra[caixa] -= item_l[item]
                    if sobra[caixa]<minimo:
                        del M[M.index(caixa)]
                    break
            i+=1
        if i==len(M):
            m+=1
            M.append(m)
            sobra[m] = L-item_l[item]
            usados[m]=[item]
    
    custo = len(list(usados.keys()))
    return usados, sobra, custo
```

```python
def FFD(l,L):
    #Ordenando os itens em ordem não-crescente
    l_ordenado = sorted(l,reverse=True)
    item_l_ordenado = {}
    #organizando o dicionário
    for item in l_ordenado:
        item_l_ordenado[l.index(item)] = item
        l.insert(l.index(item),-1) #caso haja repetição de itens com mesmo espaço isso evitará erros
        del l[l.index(item)]
    return FF(item_l_ordenado,L)
```

```python
def BFD(l,L):
    #Ordenando os itens em ordem não-crescente
    l_ordenado = sorted(l,reverse=True)
    item_l_ordenado = {}
    #organizando o dicionário
    for item in l_ordenado:
        item_l_ordenado[l.index(item)] = item
        l.insert(l.index(item),-1) #caso haja repetição de itens com mesmo espaço isso evitará erros
        del l[l.index(item)]
    return BF(item_l_ordenado,L)
```

```python
def solver(n,tipo,L):
    nomes = [f'Empacotamento_{n}_{i}.txt' for i in range(1,11)]
    resultados = []
    for nome in nomes:
        with open('instancias/'+nome,'r',encoding='utf-8') as f:
            lista = f.readlines()
            l = []
            for tamanho in lista[1:]:
                l.append(int(tamanho))
        if tipo == 'BF':
            item_l = {i:l[i] for i in range(n)}
            resultado = BF(item_l,L)
            resultados.append(resultado)
        elif tipo == 'BFD':
            resultado = BFD(l,L)
            resultados.append(resultado)
        elif tipo == "FF":
            item_l = {i:l[i] for i in range(n)}
            resultado = FF(item_l,L)
            resultados.append(resultado)
        elif tipo == 'FFD':
            resultado = FFD(l,L)
            resultados.append(resultado)
        else:
            print('Tipo Inválido!!! Tipo = BF ou BFD ou FF ou FFD')
    return resultados
```

```python
for n in [50,100,500,1000]:
    for tipo in ['FF','FFD','BF','BFD']:
        resultados = solver(n,tipo,L=100)
    print(f'Resultado para {n} itens:')
    for resultado in resultados:
        print(f'O custo final foi: {resultado[2]}')
        for b,item in zip(list(resultado[0].keys()),list(resultado[0].values())):
            print(f'Itens alocados no bin {b}: {item}')
        print('\n')
```
